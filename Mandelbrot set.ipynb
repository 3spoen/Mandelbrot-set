```python
import matplotlib.pyplot as plt
import numpy as np

def mandelbrot(x, X, y, Y):
    delta = (X-x)/500
    re, im = np.mgrid[x:X:delta, y:Y:delta]
    c = (re + 1j*im).reshape(im.shape[0], -1).T

    z = np.zeros_like(c)
    escape = np.zeros_like(np.absolute(c))
    for i in range(100):
        z = z*z + c # mandelbrot eqn
        idx = (np.absolute(z) > 4) & (escape == 0)
        escape[idx]  = i

    plt.imshow(escape, extent=(x,X,y,Y))
    
def subset_mandelbrot(x, y, D):
    return mandelbrot(x-D, x+D, y-D, y+D)
```
